```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
# Đọc dữ liệu từ file CSV
data = pd.read_csv('F:/Năm 4/HK3/Phân Tích Dữ Liệu/NhómTT/text/key-countries-pivoted_csv.csv')

# Chuyển đổi cột Date sang kiểu dữ liệu datetime
data['Date'] = pd.to_datetime(data['Date'])

# Danh sách các quốc gia bạn muốn vẽ
countries = ['China', 'US', 'United_Kingdom', 'Italy', 'France', 'Germany', 'Spain', 'Iran']

# Thiết lập kích thước của biểu đồ
plt.figure(figsize=(14, 8))

# Vẽ biểu đồ số ca nhiễm theo thời gian cho từng quốc gia trong danh sách
for country in countries:
    plt.plot(data['Date'], data[country], label=country)

# Thiết lập tiêu đề và nhãn trục
plt.title('Number of COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend()
plt.show()
```

```python
# Tính toán tỉ lệ tăng trưởng hàng ngày
growth_rate = data.set_index('Date').pct_change().dropna()

# Hiển thị vài dòng đầu của dữ liệu tỉ lệ tăng trưởng
growth_rate.head(10)
```
