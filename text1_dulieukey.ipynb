```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
# c d liu t file CSV
data = pd.read_csv('F:/Nm 4/HK3/Phn Tch D Liu/NhmTT/text/key-countries-pivoted_csv.csv')

# Chuyn i ct Date sang kiu d liu datetime
data['Date'] = pd.to_datetime(data['Date'])

# Danh sch cc quc gia bn mun v
countries = ['China', 'US', 'United_Kingdom', 'Italy', 'France', 'Germany', 'Spain', 'Iran']

# Thit lp kch thc ca biu 
plt.figure(figsize=(14, 8))

# V biu  s ca nhim theo thi gian cho tng quc gia trong danh sch
for country in countries:
    plt.plot(data['Date'], data[country], label=country)

# Thit lp tiu  v nhn trc
plt.title('Number of COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend()
plt.show()
```
