```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
# c d liu t tp CSV
data = pd.read_csv('worldwide-aggregated_csv.csv')
# Chuyn i ct Date sang kiu d liu datetime
data['Date'] = pd.to_datetime(data['Date'])

# To ct t l hi phc v t l t vong
data['Recovery Rate'] = data['Recovered'] / data['Confirmed']
data['Death Rate'] = data['Deaths'] / data['Confirmed']

# V biu  t l hi phc v t l t vong theo thi gian
plt.figure(figsize=(14, 8))
plt.plot(data['Date'], data['Recovery Rate'], label='Recovery Rate')
plt.plot(data['Date'], data['Death Rate'], label='Death Rate')
plt.xlabel('Date')
plt.ylabel('Rate')
plt.title('Recovery and Death Rates Over Time')
plt.legend()
plt.show()
```

T l tng trng hng ngy ca s ca nhim

```python
# V biu  s ca nhim, hi phc v t vong theo thi gian
plt.figure(figsize=(14, 8))
plt.plot(data['Date'], data['Confirmed'], label='Confirmed')
plt.plot(data['Date'], data['Recovered'], label='Recovered')
plt.plot(data['Date'], data['Deaths'], label='Deaths')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.title('COVID-19 Cases Over Time')
plt.legend()
plt.show()
```

 T l tng trng hng ngy ca s ca nhim

```python
# V biu  t l tng trng hng ngy ca s ca nhim
plt.figure(figsize=(14, 8))
plt.plot(data['Date'], data['Increase rate'])
plt.xlabel('Date')
plt.ylabel('Increase Rate (%)')
plt.title('Daily Increase Rate of Confirmed Cases')
plt.show()
```

```python
max_confirmed = data.loc[data['Confirmed'].idxmax()]
min_confirmed = data.loc[data['Confirmed'].idxmin()]

print("Day with highest confirmed cases:")
print(max_confirmed)

print("\nDay with lowest confirmed cases:")
print(min_confirmed)
```
