Tỉ lệ hồi phục và tử vong theo thời gian cho từng quốc gia

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
# Đọc dữ liệu từ tệp CSV
data = pd.read_csv('time-series-19-covid-combined_11_augcsv.csv')
# Chuyển đổi cột Date sang kiểu dữ liệu datetime
data['Date'] = pd.to_datetime(data['Date'])

# Lấy dữ liệu của một quốc gia cụ thể, ví dụ Afghanistan
country_data = data[data['Country/Region'] == 'Afghanistan']

# Tạo cột tỷ lệ hồi phục và tỷ lệ tử vong
country_data['Recovery Rate'] = country_data['Recovered'] / country_data['Confirmed']
country_data['Death Rate'] = country_data['Deaths'] / country_data['Confirmed']

# Vẽ biểu đồ tỷ lệ hồi phục và tỷ lệ tử vong theo thời gian
plt.figure(figsize=(14, 8))
plt.plot(country_data['Date'], country_data['Recovery Rate'], label='Recovery Rate')
plt.plot(country_data['Date'], country_data['Death Rate'], label='Death Rate')
plt.xlabel('Date')
plt.ylabel('Rate')
plt.title('Recovery and Death Rates Over Time in Afghanistan')
plt.legend()
plt.show()
```

 Số ca nhiễm, hồi phục và tử vong theo thời gian cho từng quốc gia

```python
# Vẽ biểu đồ số ca nhiễm, hồi phục và tử vong theo thời gian
plt.figure(figsize=(14, 8))
plt.plot(country_data['Date'], country_data['Confirmed'], label='Confirmed')
plt.plot(country_data['Date'], country_data['Recovered'], label='Recovered')
plt.plot(country_data['Date'], country_data['Deaths'], label='Deaths')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.title('COVID-19 Cases Over Time in Afghanistan')
plt.legend()
plt.show()
```

Tỉ lệ tăng trưởng hàng ngày của số ca nhiễm cho từng quốc gia

```python
# Tính toán tỉ lệ tăng trưởng hàng ngày
country_data['Daily Growth Rate'] = country_data['Confirmed'].pct_change() * 100

# Vẽ biểu đồ tỉ lệ tăng trưởng hàng ngày của số ca nhiễm
plt.figure(figsize=(14, 8))
plt.plot(country_data['Date'], country_data['Daily Growth Rate'])
plt.xlabel('Date')
plt.ylabel('Daily Growth Rate (%)')
plt.title('Daily Growth Rate of Confirmed Cases in Afghanistan')
plt.show()
```

Số ca nhiễm, hồi phục và tử vong tổng hợp theo thời gian

```python
# Tổng hợp số ca nhiễm, hồi phục và tử vong theo thời gian cho toàn thế giới
global_data = data.groupby('Date').sum().reset_index()

# Vẽ biểu đồ số ca nhiễm, hồi phục và tử vong theo thời gian
plt.figure(figsize=(14, 8))
plt.plot(global_data['Date'], global_data['Confirmed'], label='Confirmed')
plt.plot(global_data['Date'], global_data['Recovered'], label='Recovered')
plt.plot(global_data['Date'], global_data['Deaths'], label='Deaths')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.title('Global COVID-19 Cases Over Time')
plt.legend()
plt.show()
```
